```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x = np.arange(0.0, 2, 0.1)
y1 = 10000*np.exp(-6 * x)
y2 = 10000*np.exp(-4 * x)
y3 = 10000*np.exp(-1.5 * x)
y4 = 10000*np.exp(-7.5 * x)

fig, (ax1) = plt.subplots(1, 1, sharex=True)

ax1.fill_between(x, 0, y1, color='blue')
ax1.fill_between(x, y1, y2, color='green')
ax1.fill_between(x, y2, y3, color='yellow')
ax1.fill_between(x,y3,10000, color='red')
ax1.set_ylabel('Social Media Volume')
ax1.set_xlabel('Time')
```
